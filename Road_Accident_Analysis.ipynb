```python
import pandas as pd

file_path = "/content/drive/My Drive/Road Accident Data.csv"  # Change this to your actual file name
df = pd.read_csv(file_path)

df.head()  # Display the first 5 rows
```

```python
print("Total number of accidents:", len(df))
```

```python
df['Year'] = pd.to_datetime(df['Accident Date']).dt.year

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.countplot(x='Year', data=df, palette='coolwarm')
plt.title("Accidents Per Year")
plt.show()
```
